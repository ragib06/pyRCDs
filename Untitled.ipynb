```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_path([1,2,3,4,5])
G.add_path([2,4,7,9])
G.add_path([1,3,5,8])
pos = nx.nx_agraph.graphviz_layout(G)
nx.draw(G, pos=pos)
plt.draw() 
labels = nx.draw_networkx_labels(G, pos=pos)
plt.show()
```
